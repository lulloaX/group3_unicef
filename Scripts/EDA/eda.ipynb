## Aggregate

```python
df_ucayali.drop(labels = ["CENTRO_SALUD"], axis =1, inplace =True)
```

```python
df_result = df_ucayali.groupby(['VACUNA', 'ANIO', 'MES', "PROVINCIA", "DISTRITO"]).agg({
    'RANGO_EDAD': 'first',
    'CANTIDAD DE PERSONAL': 'sum', 
    'POBLACION': 'first',
    'PRESUPUESTO':'first',
    'CANTIDAD':'sum'
}).reset_index()
df_result
```

```python
df_result_2 = df_ucayali.groupby(['VACUNA', 'ANIO', 'MES','RANGO_EDAD', 'POBLACION',"PROVINCIA", "DISTRITO" ,"PRESUPUESTO"]).agg({
    'CANTIDAD DE PERSONAL': 'sum',
    'CANTIDAD':'sum'
}).reset_index()
df_result_2
```

```python
df_result.to_csv("../Dataset/Ucayali/df_result.csv", index = False)
df_result_2.to_csv("../Dataset/Ucayali/df_result_2.csv", index = False)
```

```python
df_result_2.merge(df_result, indicator=True, how='left').loc[lambda x: x['_merge'] == 'left_only'].drop('_merge', axis=1)["VACUNA"]
```
