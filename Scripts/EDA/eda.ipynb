## Aggregate

```python
df_ucayali.drop(labels = ["CENTRO_SALUD"], axis =1, inplace =True)
```

```python
df_result = df_ucayali.groupby(['VACUNA', 'ANIO', 'MES', "PROVINCIA", "DISTRITO"]).agg({
    'RANGO_EDAD': 'first',
    'CANTIDAD DE PERSONAL': 'sum', 
    'POBLACION': 'first',
    'PRESUPUESTO':'first',
    'CANTIDAD':'sum'
}).reset_index()
df_result
```

## Plots

```python
df_plots = pd.read_csv("../../Dataset/Ucayali/df_datos_sorted_Shifted_log.csv")
df_plots.dropna(subset=["CANTIDAD"], inplace = True)
df_plots
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
for item, df_group in df_plots.groupby(["PROVINCIA","VACUNA"]):
    df_group["TARGET"] = np.exp(df_group['CANTIDAD']) - 0.001
    plt.hist(df_group['TARGET'], bins=20, color='skyblue', edgecolor='black')
    plt.xlabel('Cantidad de personas vacunadas')
    plt.ylabel('Frecuencia')
    plt.title(f'Histograma del target CANTIDAD de personas vacunadas: {item}')
    plt.show()
```
